```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
from calibration import lifetime_to_temperature_201112, pict_calibration
from scipy.optimize import curve_fit
import seaborn as sns
from PIL import Image
import scipy as sp
from matplotlib.backends.backend_pdf import PdfPages
```

```python
df = pd.read_csv('../../Fig4/Fig4b/Fig4b_avetmp.csv', header = None)
time_500msec = df[0].dropna()
deltaT_500msec = df[1].dropna()
time_5sec = df[2].dropna()
deltaT_5sec = df[3].dropna()
```

```python
fig = plt.figure(figsize = (20, 20))
plt.tick_params(axis='both', direction = "in", labelsize = 24)
plt.plot(time_500msec, deltaT_500msec, color = 'black')
plt.plot(time_5sec, deltaT_5sec, color = 'red')
plt.ylim(-0.4, 1.2)
plt.xlim(-6, 10)

'''
#1˚Cを示すバーを挿入
bar_500msec = [100 - 20/deltaT_500msec[10]*i/100 for i in range(100)]
bar_500msec_pos = [3.4 for i in range(len(bar_500msec))]
plt.plot(bar_500msec_pos, bar_500msec, linewidth = 4, color = 'black', linestyle = '-')

#1˚Cを示すバーを挿入
bar_5sec = [100 - 20/deltaT_5sec[10]*i/100 for i in range(100)]
bar_5sec_pos = [3.6 for i in range(len(bar_500msec))]
plt.plot(bar_5sec_pos, bar_5sec, linewidth = 4, color = 'red', linestyle = '-')
'''

plt.show()



pp = PdfPages('../../Program/Uneditied_figure/Fig4b/ave_tmp.pdf')
pp.savefig(fig)
pp.close()
```

```python
fig = plt.figure(figsize = (20, 20))
plt.tick_params(axis='both', direction = "in", labelsize = 24)
plt.plot(time_500msec[10:35], 100*deltaT_500msec[10:35]/deltaT_500msec[10], color = 'black', linewidth = 0, marker = 'o', markersize = 20)
plt.plot(time_5sec[40:65], 100*deltaT_5sec[40:65]/deltaT_5sec[40], color = 'red', linewidth = 0, marker = 'o', markersize = 20)
plt.ylim(-35, 110)
plt.xlim(-0.2, 3.5)


#温度緩和の時定数の計算
def relaxation(time, ratio, t1, t2):
    return ratio*np.exp(-time/t1) + (1-ratio)*np.exp(-time/t2)
result_500msec = curve_fit(relaxation, time_500msec[10:35], deltaT_500msec[10:35]/deltaT_500msec[10])[0]
result_5sec = curve_fit(relaxation, time_5sec[40:65], deltaT_5sec[40:65]/deltaT_5sec[40])[0]

plt.plot(time_500msec[10:35], relaxation(np.ravel(time_500msec[10:35]), result_500msec[0], result_500msec[1], result_500msec[2])*100, color = 'black')
plt.plot(time_5sec[40:65], relaxation(np.ravel(time_5sec[40:65]), result_5sec[0], result_5sec[1], result_5sec[2])*100, color = 'red')



#1˚Cを示すバーを挿入
bar_500msec = [100 - 20/deltaT_500msec[10]*i/100 for i in range(100)]
bar_500msec_pos = [2.6 for i in range(len(bar_500msec))]
plt.plot(bar_500msec_pos, bar_500msec, linewidth = 4, color = 'black', linestyle = '-')

#1˚Cを示すバーを挿入
bar_5sec = [100 - 20/deltaT_5sec[10]*i/100 for i in range(100)]
bar_5sec_pos = [2.8 for i in range(len(bar_500msec))]
plt.plot(bar_5sec_pos, bar_5sec, linewidth = 4, color = 'red', linestyle = '-')

plt.show()

pp = PdfPages('../../Program/Uneditied_figure/Fig4b/relaxation.pdf')
pp.savefig(fig)
pp.close()
```
